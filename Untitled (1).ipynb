```python
import requests
from bs4 import BeautifulSoup

# Loop through the first link in df['link']
dct = {'link':[],'ref_link':[]}
for x in df['link'][50:]:
    # Request the base page
    response = requests.get(x)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all matching links on the base page
    links = soup.find_all('a', class_='newsreleaseconsolidatelink')
    
    if links:
        # Iterate through paginated pages
        for i in range(1, 300):
            url = f"{x}?page={i}&pagesize=100"
            print(i)
            response = requests.get(url)  # Corrected to request the paginated URL
            html_content = response.text
            soup = BeautifulSoup(html_content, 'html.parser')

            # Extract links from the paginated page
            paginated_links = soup.find_all('a', class_='newsreleaseconsolidatelink')

            if paginated_links:
                for link in paginated_links:
                    dct['link'].append(link['href'])
                    dct['ref_link'].append(x)
            else:
                break  # Stop looping if no links found on a page
df_new_data = pd.DataFrame(dct)
# Saving File To New Loc:
df_new_data.to_csv("data_Dt-14-04-25.csv", index=False, encoding='utf-8')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Create an empty list to store the extracted links
extracted_links = []

# Loop through the first link in df['link']
for x in df['link'][:50]:
    # Request the base page
    response = requests.get(x)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all matching links on the base page
    links = soup.find_all('a', class_='newsreleaseconsolidatelink')

    if links:
        # Iterate through paginated pages
        for i in range(1, 300):
            url = f"{x}?page={i}&pagesize=100"
            response = requests.get(url)  # Request the paginated URL
            html_content = response.text
            soup = BeautifulSoup(html_content, 'html.parser')

            # Extract links from the paginated page
            paginated_links = soup.find_all('a', class_='newsreleaseconsolidatelink')

            if paginated_links:
                for link in paginated_links:
                    # Append the link to the list
                    extracted_links.append(link['href'])
            else:
                break  # Stop looping if no links found on a page
    # df_links = pd.DataFrame({'link': extracted_links})
    # df_links.to_csv('extracted_links.csv', index=False)

# Create a Pandas DataFrame from the list of links
df_links = pd.DataFrame({'link': extracted_links})

# Export the DataFrame to a CSV file
df_links.to_csv('extracted_links.csv', index=False)

print("Links have been saved to 'extracted_links.csv'")
```
